# Redes Neuronales Artificiales

## Dataset: Sonar - https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks)

```python
# Import data handling libraries
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

## Semilla para random
Este valor de semilla es utilizado para que el codigo siempre genere el mismo resultado en procesos aleatorios

```python
# fix random seed for reproducibility
seed = 1001
np.random.seed(seed)
```

## Lectura del dataset
Se carga el dataset usando Pandas, creacion del vector "X" de variables independientes y el vector columna target "y" (ambos vectores de Numpy)

```python
# Read dataset
df = pd.read_csv('sonar.csv', header=0)
X = df.iloc[:, 0:60].values
y = df.iloc[:, 60].values

df.head(10)
```

## Transformación de variables
El vector objetivo "y" es transformado para comvertir sus valores de "R" por el entero 1 y el valor de "M" por el valor de 0

```python
# Encode target feature to integer
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
y[91:101]
```

## Division del dataset
Se utiliza 80% de los datos para entrenamiento y 20% para pruebas de presicion en la predicción

```python
# Split dataset 80% for training vs 20% for test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Creación de la red neuronal artificial
### Estructura
La red neuronal tiene 60 unidades de entrada, 60 unidades en la capa oculta y 1 unidad en el capa de salida. Los pesos sinápticos son escogidos al azar siguiendo una distribución normal (Valores entre -3 a 3). Se utiliza el modelo estocastico secuencial donde cada neurona sigue la formula "output = activation(dot(input, kernel) + bias)". Las funciones de activación utilizadas en las capas es la función sigmoidal debido a la naturaleza binaria del problema.

```python
# Create ANN (60 imput units, 60 hidden units, 1 output unit) with weights following normal distribution
from keras.models import Sequential # Stochastic
from keras.layers import Dense # output = activation(dot(input, kernel) + bias)
binary_classifier = Sequential()
binary_classifier.add(Dense(60, input_dim=60, kernel_initializer='normal', activation='sigmoid'))
binary_classifier.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
```

## Compilado y entrenamiento
Se utiliza la entropia cruzada binaria como funcion de costos y el optimizador Adam (adaptive momentum) de nuevo por la naturaleza binaria del problema. Adam utiliza una actualización del gradiente mediante movimiento exponencial promedio de los gradientes pasados.

El entrenamiento utiliza 100 epochs (recorridos al total del dataset) con un batch de tamaño 5 (numero de observaciones antes de actualización de los pesos sinápticos)

```python
# Compile using adaptive momentum optimizer with binary cross entropy as cost funtion
binary_classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
binary_classifier.fit(X_train, y_train, batch_size=5, epochs = 100)
```

## Predicción
Se calcula la predicción para el dataset de prueba, debido a que la red neuronal retorna probabilidades entre 0 y 1, los valors son llevados a piso y techo a partir de 0.5

```python
# Create and transform probabilistic predictions
y_pred = binary_classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

## Matriz de confusion
Se calcula la matriz de confusion para evaluar la presición y la cantidad de errores tipo 1 y tipo 2 del modelo.

```python
# Create confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

## Calculo de la presición
Se calcula la presición final de la red neuronal a partir de la matriz de confusión.

```python
# Compute accuracy
acc = (cm[0,0] + cm[1,1]) / len(y_test)
print('Accuracy: {0}'.format(acc))
```
